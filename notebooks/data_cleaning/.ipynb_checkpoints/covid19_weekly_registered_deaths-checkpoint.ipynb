# Covid-19 Deaths Registered Weekly

Data downloaded from [weeklyprovisionalfiguresondeathsregisteredinenglandandwales](https://www.ons.gov.uk/peoplepopulationandcommunity/birthsdeathsandmarriages/deaths/datasets/weeklyprovisionalfiguresondeathsregisteredinenglandandwales)

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

### Load data

```python
df = pd.read_excel('../../data/publishedweek512020corrected.xlsx',
                   sheet_name='Weekly figures 2020',
                   header=5,
                   skipfooter=13)

df = df.drop(df.columns[-2:], axis=1)
```

```python
df.head()
```

```python
df.tail()
```

I'll start by tidying the dates up. We don't need the time.

```python
for i in df.columns[2:]:
    df.rename(columns={i: pd.datetime.date(i)}, inplace=True)
```

```python
df.head(10)
```

```python
df_sum = df.iloc[:13]
df_sum.rename(columns={'Week ended': 'Summary'}, inplace=True)
df_sum['Summary'][11] = df_sum['Unnamed: 1'][11]
df_sum['Summary'][12] = df_sum['Unnamed: 1'][12]
df_sum.set_index('Summary', inplace=True)
df_sum.drop(columns=['Unnamed: 1'], inplace=True)
df_sum.dropna(axis=0, how='all', inplace=True)
df_sum
```

```python
df_reg = df.iloc[80:]
df_reg.rename(columns={'Unnamed: 1': 'Region'}, inplace=True)
df_reg.set_index('Region', inplace=True)
df_reg.drop(columns='Week ended', inplace=True)

df_reg
```

```python
df_age = df.iloc[15:79]
df_age.drop(columns='Week ended', inplace=True)
df_age.drop([df_age.index[20], df_age.index[21], df_age.index[42], df_age.index[43]], inplace=True)

sex = (['Persons']*len(df_age['Unnamed: 1'].unique()) +
       ['Males']*len(df_age['Unnamed: 1'].unique()) +
       ['Females']*len(df_age['Unnamed: 1'].unique()))

index = list(zip(*[sex, df_age['Unnamed: 1']]))
index = pd.MultiIndex.from_tuples(tuple(zip(sex, df_age['Unnamed: 1'])), names=["Sex", "Age"])

df_age.drop(columns='Unnamed: 1', inplace=True)

df_age.set_index(index, inplace=True)



df_age

#df_age.rename(columns={'Unnamed: 1': 'Age'}, inplace=True)
#df_age.drop(columns='Week ended', inplace=True)
#df_age.set_index('Age', inplace=True)
#df_age
```
